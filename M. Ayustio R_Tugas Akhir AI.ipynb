```python
import matplotlib.pyplot as plt
import networkx as nx

# 1. Definisi graf arah dari Rajabasa ke TPA
graf_tpa = {
    'Rumah': ['TPA Untung', 'TPA PasarRajabasa'],
    'TPA Untung': ['TPA WayHalim'],
    'TPA PasarRajabasa': ['TPA Kedaton'],
    'TPA WayHalim': ['TPA KpBaru'],
    'TPA Kedaton': ['TPA KpBaru'],
    'TPA KpBaru': []
}

# 2. Komposisi sampah per TPA
komposisi_sampah = {
    'TPA Untung': {'organik': 70, 'plastik': 20, 'logam': 10},
    'TPA PasarRajabasa': {'organik': 40, 'plastik': 50, 'logam': 10},
    'TPA WayHalim': {'organik': 25, 'plastik': 60, 'logam': 15},
    'TPA Kedaton': {'organik': 55, 'plastik': 30, 'logam': 15},
    'TPA KpBaru': {'organik': 20, 'plastik': 70, 'logam': 10}
}

# 3. Buat graf terarah
G = nx.DiGraph()
for titik, tetangga in graf_tpa.items():
    for tujuan in tetangga:
        G.add_edge(titik, tujuan)

# 4. Buat tata letak node
pos = nx.spring_layout(G, seed=42)

# 5. Gambar node dan edge
plt.figure(figsize=(12, 8))
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2500)
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# 6. Tambahkan label komposisi sampah di bawah setiap node TPA
for node in G.nodes:
    if node in komposisi_sampah:
        data = komposisi_sampah[node]
        label = f"O:{data['organik']}% P:{data['plastik']}% L:{data['logam']}%"
        x, y = pos[node]
        plt.text(x, y - 0.08, label, fontsize=9, ha='center', color='darkgreen')

# 7. Tampilan akhir
plt.title("Graf TPA Sekitar Rajabasa dan Komposisi Sampah", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Komposisi sampah per TPA
komposisi_sampah = {
    'TPA Untung': {'organik': 70, 'plastik': 20, 'logam': 10},
    'TPA PasarRajabasa': {'organik': 40, 'plastik': 50, 'logam': 10},
    'TPA WayHalim': {'organik': 25, 'plastik': 60, 'logam': 15},
    'TPA Kedaton': {'organik': 55, 'plastik': 30, 'logam': 15},
    'TPA KpBaru': {'organik': 20, 'plastik': 70, 'logam': 10}
}

# Ambil input dari user
jenis_dicari = input("Masukkan jenis sampah yang ingin dicari (organik/plastik/logam): ").lower()

# Validasi input
if jenis_dicari not in ['organik', 'plastik', 'logam']:
    print("Jenis sampah tidak dikenali. Pilih antara: organik, plastik, atau logam.")
else:
    # Cari TPA dengan nilai tertinggi untuk jenis yang dicari
    max_tpa = None
    max_persen = -1

    for tpa, komposisi in komposisi_sampah.items():
        persen = komposisi.get(jenis_dicari, 0)
        if persen > max_persen:
            max_persen = persen
            max_tpa = tpa

    # Tampilkan hasil
    print(f"\n TPA dengan kandungan {jenis_dicari} tertinggi:")
    print(f" {max_tpa} dengan {max_persen}% {jenis_dicari}")
    print("\n Komposisi lengkap:")
    for jenis, nilai in komposisi_sampah[max_tpa].items():
        print(f"  - {jenis.capitalize()}: {nilai}%")
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# 1. Komposisi sampah per TPA
komposisi_sampah = {
    'TPA Untung': {'organik': 70, 'plastik': 20, 'logam': 10},
    'TPA PasarRajabasa': {'organik': 40, 'plastik': 50, 'logam': 10},
    'TPA WayHalim': {'organik': 25, 'plastik': 60, 'logam': 15},
    'TPA Kedaton': {'organik': 55, 'plastik': 30, 'logam': 15},
    'TPA KpBaru': {'organik': 20, 'plastik': 70, 'logam': 10}
}

# 2. Graf arah dari Rajabasa
graf_tpa = {
    'Rumah': ['TPA Untung', 'TPA PasarRajabasa'],
    'TPA Untung': ['TPA WayHalim'],
    'TPA PasarRajabasa': ['TPA Kedaton'],
    'TPA WayHalim': ['TPA KpBaru'],
    'TPA Kedaton': ['TPA KpBaru'],
    'TPA KpBaru': []
}

# 3. Ambil input jenis sampah
jenis_dicari = input("Masukkan jenis sampah (organik/plastik/logam): ").lower()

# 4. Validasi input
if jenis_dicari not in ['organik', 'plastik', 'logam']:
    print("Jenis tidak valid. Pilih: organik, plastik, logam.")
else:
    # Cari TPA dengan jumlah tertinggi
    target_tpa = max(komposisi_sampah.items(), key=lambda item: item[1][jenis_dicari])[0]

    print(f"\nTPA dengan {jenis_dicari} terbanyak adalah {target_tpa}")
    print("Komposisinya:", komposisi_sampah[target_tpa])

    # 5. Buat graf dan highlight target
    G = nx.DiGraph()
    for titik, tetangga in graf_tpa.items():
        for tujuan in tetangga:
            G.add_edge(titik, tujuan)

    pos = nx.spring_layout(G, seed=42)
    warna_node = ['red' if node == target_tpa else 'lightblue' for node in G.nodes]

    plt.figure(figsize=(12, 8))
    nx.draw_networkx_nodes(G, pos, node_color=warna_node, node_size=2500)
    nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray')
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    # Tampilkan label komposisi sampah di bawah node
    for node in G.nodes:
        if node in komposisi_sampah:
            data = komposisi_sampah[node]
            label = f"O:{data['organik']}% P:{data['plastik']}% L:{data['logam']}%"
            x, y = pos[node]
            plt.text(x, y - 0.08, label, fontsize=9, ha='center', color='darkgreen')

    plt.title(f"Graf TPA Rajabasa  Fokus: {jenis_dicari.upper()} terbanyak", fontsize=14)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```
